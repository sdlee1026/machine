```python
import warnings
warnings.filterwarnings(action='ignore')
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)
# 데이터의 shuffle : boolean, optional (default=True) 매개변수 사용
# shuffle=false인 경우, stratify는 'None'이어야 한다.
# 데이터가 array와 같은 경우, stratify : array-like or None (default=None) 매개변수 사용
# stratify를 default로 사용하지 않는 경우 (default가 True, False 가 아닌 None임)
# 데이터는 계층화된 방식으로 분할되고, 그것이 클래스 라벨로 사용될 것, 지금의 경우 클래스 라벨을 cancer.target으로 지정
# shuffle default가 True기 때문에 자동으로 데이터들은 순서가 섞여서 나누어짐
```
