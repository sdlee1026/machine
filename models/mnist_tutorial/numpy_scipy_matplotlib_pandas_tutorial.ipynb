```python
from scipy import sparse
# 과학 계산용 함수를 모아놓은 파이썬 패키지 (scikit-learn 알고리즘은 Scipy의 여러함수를 사용_!Scipy.sparse!)
# 희소 행렬 기능을 제공
```

```python
# 대각선 원소는 1, 나머지는 0인 2차원 희소 행렬
eye = np.eye(8,9)
```

```python
eye
```

```python
# Numpy 배열을 CSR 포맷의 Scipy 희박 행렬로 변환
# 0이 아닌 원소만 저장됨
sparse_matrix = sparse.csr_matrix(eye)
# CSR_Compressed Sparse Row, 행의 인덱스를 압축하여 저장
```

```python
print(sparse_matrix)
# index 쌍 + values
```

```python
# 보통 희소 행렬을 0이 모두 채워진 2차원 배열로부터 만들지 않음_메모리 너무 많이 먹기 때문에
# 희소 행렬을 직접만드는.. COO포멧을 이용함_Coordinate format ,데이터가 놓일 행렬의 위치를 별도의 매개 변수로 전달
```

```python
data = np.ones(4)
print(data)
```

```python
row_indices = np.arange(4)
col_indices = np.arange(4)
```

```python
print('row_indices : ', row_indices)
print('col_indices : ', col_indices)
```

```python
eye_coo = sparse.coo_matrix((data, (row_indices,col_indices)))
```

```python
print(eye_coo)
# 표현은 csr과 coo는 동일하게 보여지지만, coo는 값을 직접 별도의 매개변수로 전달하고, csr는 행의 인덱스를 압축해서 저장하는 차이가 있다
```
